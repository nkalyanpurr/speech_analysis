```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import *
from collections import Counter
import re
import string

with open('debate.txt','r') as f:
    debate = f.read()

def Obama(str):
    return re.sub('^OBAMA:', '', str)

def Lehrer(str):
    return re.sub('^LEHRER:', '', str)

def Romney(str):
    return re.sub('^ROMNEY:', '', str)
paragraphs = [p for p in debate.split('\n') if p]
Obama1 = "OBAMA:"
Lehrer1 = "LEHRER:"
Romney1 = "ROMNEY:"
v = 0
#concatenate statements made by the same speaker 
for paragraph in paragraphs:
    result1 = re.match("^OBAMA:", paragraph)
    result2 = re.match("^LEHRER:", paragraph)
    result3 = re.match("^ROMNEY:", paragraph)
    if result1:
        Obama1 += Obama(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
        v = 1

    elif result2:
        Lehrer1 += Lehrer(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
        v = 2

    elif result3:
        Romney1 += Romney(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
        v = 3
    else:
        if (v == 1):
            Obama1 += Obama(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
        if (v == 2):
            Lehrer1 += Lehrer(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
        if (v == 3):
            Romney1 += Romney(re.sub('(\()[A-Za-z]*(\))', '', paragraph))
#displaying the results
print(Obama1)
print("\n***")
print(Lehrer1)
print("\n***")
print(Romney1)
print("\n***")

```

```python
#Removing the punctuations and capitalization
translator = str.maketrans('', '', string.punctuation)
Obama1 = Obama1.translate(translator).lower()
Lehrer1 = Lehrer1.translate(translator).lower()
Romney1 = Romney1.translate(translator).lower()
print(Obama1)
print("\n\n")
print(Lehrer1)
print("\n\n")
print(Romney1)
```

```python
#Removing stopwords and tokenizing the words.
stop_words = set(stopwords.words('english')) 
word_tokens = word_tokenize(Obama1)
word_tokens1 = word_tokenize(Lehrer1)
word_tokens2 = word_tokenize(Romney1)
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
filtered_sentence = [] 
filtered_sentence1 = [w for w in word_tokens if not w in stop_words] 
filtered_sentence1 = [] 
filtered_sentence2= [w for w in word_tokens if not w in stop_words] 
filtered_sentence2 = [] 
 
for w in word_tokens: 
    if w not in stop_words: 
        filtered_sentence.append(w) 

print(filtered_sentence)

for w in word_tokens1: 
    if w not in stop_words: 
        filtered_sentence1.append(w) 

print(filtered_sentence1)

for w in word_tokens2: 
    if w not in stop_words: 
        filtered_sentence2.append(w) 

print(filtered_sentence2)
```

```python
#Applying the stemmers
ps = PorterStemmer()
psobama = [ps.stem(w) for w in filtered_sentence]
pslehrer = [ps.stem(w) for w in filtered_sentence1]
psromney = [ps.stem(w) for w in filtered_sentence2]
print(psobama)
print(pslehrer)
print(psromney)
sb = SnowballStemmer("english")
sbobama = [sb.stem(w) for w in filtered_sentence]
sblehrer = [sb.stem(w) for w in filtered_sentence1]
sbromney = [sb.stem(w) for w in filtered_sentence2]
print(sbobama)
print(sblehrer)
print(sbromney)
ls = LancasterStemmer()
lsobama = [ls.stem(w) for w in filtered_sentence]
lslehrer = [ls.stem(w) for w in filtered_sentence1]
lsromney = [ls.stem(w) for w in filtered_sentence2]
print(lsobama)
print(lslehrer)
print(lsromney)
```

```python
#Outputting 10 frequent words from each stemmer
from collections import Counter

Counter(psobama).most_common()
d = {}
for word in psobama: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(PorterStemmer) Obama:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1

d = {}
for word in pslehrer: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(PorterStemmer) Lehrer:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1

d = {}
for word in psromney: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(PorterStemmer) Romney:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1
        

d = {}
for word in sbobama: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(Snowball) Obama:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1

d = {}
for word in sblehrer: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(Snowball) Lehrer:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1
        

d = {}
for word in sbromney: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(Snowball) Romney:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1
        

d = {}
for word in lsobama: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(LancasterStemmer) Obama:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1

d = {}
for word in lslehrer: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(LancasterStemmer) Lehrer:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1

d = {}
for word in lsromney: 
    d[word] = d.get(word, 0) + 1
word_freq = []
for key, value in d.items():
    word_freq.append((value, key))
word_freq.sort(reverse=True) 
print("(LancasterStemmer) Romney:")
counter = 0
for wf in word_freq:
    if counter <10:
        print (wf)
    
        counter += 1
        

```

The Porter Stemmer and Snowball are almost similar but in some cases snowball overpowers  and is more advanced compared to Porterstemmer. Whereas, Lancaster is more precise with more number of frequency counts.

```python
#Displaying 10 most frequent words used by each user
with open("positive.txt", 'r') as f:
    positive = f.read().split('\n')
pspositive = [ps.stem(w) for w in positive]
posobama = (set(psobama) & set(pspositive))
poslehrer = (set(pslehrer) & set(pspositive))
posromney = (set(psromney) & set(pspositive))
print("Positive words from Obama:")
n = Counter()
cnt = Counter(psobama)
for w in posobama:
    n[w]=cnt[w]
print(Counter(n).most_common(10))
print("Positive words from Lehrer:")
b = Counter()
cnt = Counter(pslehrer)
for w in poslehrer:
    b[w]=cnt[w]
print(Counter(b).most_common(10))
print("Positive words from Romney:")
c = Counter()
cnt = Counter(psromney)
for w in posromney:
    c[w]=cnt[w]
print(Counter(c).most_common(10))


```
